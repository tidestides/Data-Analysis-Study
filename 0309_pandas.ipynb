07-1 열과 피벗

  1. melt 메서드 - 지정한 열의 데이터를 모두 행으로 정렬
* id_vars - 위치를 그대로 유지할 열 이름 설정
* value_vars - 행으로 위치를 변경할 열의 이름 지정
* var_name - value_vars로 위치를 변경한 열의 이름 지정
* var_name - 위치를 변경한 열의 데이터를 저장한 열의 이름 지정

<1개의 열 고정하여 피벗>

```python
#불러오기
import pandas as pd
pew = pd.read_csv('pew.csv')
```

```python
#6개의 열만 출력
pew.iloc[:, 0:6]
```

```python
#religion 열을 고정하여 피벗하기
#종교 컬럼의 열을 그대로 유지하되, 나머지 열은 새롭게 정리됨
pew_long = pd.melt(pew, id_vars = 'religion')

pew_long
```

```python
#컬럼명 지정해 주기!
pew_long = pd.melt(pew, id_vars = 'religion', var_name = 'income', value_name ='count')

pew_long
```

<2개 이상의 열 고정해 피벗>

```python
#불러와서 최상위 행 5개 열 16개만 출력

billboard = pd.read_csv('billboard.csv')
billboard.iloc[0:5, 0:16]
```

```python
#id_vars에 고정할 열 이름 설정
#var_name으로 피벗할 열의 새로운 열 이름 부여
#var_name으로 위치를 변경한 열의 데이터를 저장한 열의 이름 지정
billboard_long = pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'], var_name = 'week',
                        value_name='rating')

billboard_long
```

2. 피벗

```python
#불러와서 상위 5개 행까지만 보기
weather = pd.read_csv('weather.csv')

weather.iloc[:5,:]
```

```python
#id_vars에 적힌 네 개 열을 고정해 피벗

weather_melt = pd.melt(weather, id_vars=['id','year','month','element'],
                      var_name='day',value_name='temp')

weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(index=['id','year','month','day'],
                                       columns = 'element',
                                       values ='temp')

weather_tidy.reset_index() # 원본 데이터를 변경하고 싶은 경우 .reset_index(inplace = True)
```

<중복 데이터 처리 >

```python
#불러오기

billboard = pd.read_csv('billboard.csv')

billboard
```

```python
#피벗
billboard_long = pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'],
                        var_name = 'week', value_name='rating')

billboard_long
```

```python
#선생님 추천 
billboard_long[billboard_long.track=='Loser']
```

```python
billboard_songs =billboard_long[['year','artist','track','time']]
```

```python
#drop_duplicates() - 중복 데이터 날리기
billboard_songs.drop_duplicates()
```

```python
billboard_songs['id'] = range(len(billboard_songs))
```

```python
billboard_long
```

```python
#중복값 삭제
billboard_ratings = billboard_long.merge(billboard_songs, on=['year','artist','track','time'])

billboard_ratings.drop_duplicates()
```

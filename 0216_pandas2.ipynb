matplotlib 사용하기

```python
fig = plt.figure() #도화지 그리기
axes1 = fig.add_subplot(1,1,1) #도화지의 사이즈 설정
axes1.hist(tips['total_bill'], bins=10) #뽑을 컬럼, 표현할 막대의 수
axes1.set_title('Histogram of Tatal_Bill', fontsize=10) #표 타이틀명과 글씨크기 설정
axes1.set_xlabel('Frequency') #x라벨 이름 설정
axes1.set_ylabel('Total_bill') #Y라벨 이름 설정
```

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('scatter', fontsize=10)
axes1.set_xlabel('total bill')
axes1.set_ylabel('tip')
```

##박스플롯(상자수염그림)의 구조 이해
- 제1사분위수, 제2사분위수, 제3사분위수는 각각 상위 25%, 50%, 75%를 의미한다.
- IQR: 상위 25%의 관측값과 하위 25%의 관측값을 제외한 중앙 50%의 관측값의 퍼진 정도

- 박스의 범위는 IQR을 의미 (전체 분포 범위의 가운데 50%)
- 박스 안 선은 데이터 중앙값 의미
- 수염은 특이치를 제외한 데이터 하위 25%와 상위 25%를 의미
- 아웃라이어: 상하위 25%보다 1.5배 크거나 작은 값

##왜 박스플롯을 쓰나?
- 이상점을 처리하는 데 뛰어나고, 데이터 전체의 편포 여부를 확인하는 것이 쉬움
- 데이터의 집단 간 비교, 데이터의 타당도(validity)를 체크하는 것도 용이

```python
#boxplot

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
              tips[tips['sex']=='Male']['tip']],
             labels = ['Female','Male'])
axes1.set_title('boxplot', fontsize=10)
axes1.set_xlabel('Sex')
axes1.set_ylabel('tip')
```
